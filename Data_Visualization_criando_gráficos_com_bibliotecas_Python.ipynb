```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the immigration data again
df_immigration = pd.read_csv('/canadian_immegration_data.csv')
df_immigration.set_index('Country', inplace=True)  # Set the index to 'Country'

fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Use df_immigration instead of df
axs[0, 0].plot(df_immigration.loc['Brazil', anos])  # Note: 'Brazil' instead of 'Brasil'
axs[0, 0].set_title('Brazil')

axs[0, 1].plot(df_immigration.loc['Colombia', anos])
axs[0, 1].set_title('Colombia')

axs[1, 0].plot(df_immigration.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))

# Nome do paÃ­s corrigido de 'Brasil' para 'Brazil''
axs[0,0].plot(df_immigration.loc['Brazil', anos])
axs[0,0].set_title('Brazil')

# Nome do paÃ­s corrigido de 'ColÃ´mbia' para 'Colombia''
axs[0,1].plot(df_immigration.loc['Colombia', anos])
axs[0,1].set_title('Colombia')

axs[1,0].plot(df_immigration.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df_immigration.loc['Peru', anos])
axs[1,1].set_title('Peru')

plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

axs[0,0].plot(df_immigration.loc['Brazil', anos])
axs[0,0].set_title('Brazil')

axs[0,1].plot(df_immigration.loc['Colombia', anos])
axs[0,1].set_title('Colombia')

axs[1,0].plot(df_immigration.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df_immigration.loc['Peru', anos])
axs[1,1].set_title('Peru')

plt.show()
```

```python
for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
```
